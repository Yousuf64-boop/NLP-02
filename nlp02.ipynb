```python
documents = [
    'people watch campusx',
    'campusx watch campusx',
    'people write comment',
    'campusx write comment'
]
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
vocab = list(set(" ".join(documents).split()))

word_to_index = {word: i for i, word in enumerate(vocab)}
```

```python
def one_hot(word):
    vector = np.zeros(len(vocab))
    vector[word_to_index[word]] = 1
    return vector
```

```python
print('Vocabulary:', vocab)
print("One_hot for 'people':", one_hot('people'))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

X = cv.fit_transform(documents)
print("Vocabulary:", cv.get_feature_names_out())
print("Bow Matrix:\n", X.toarray())
```

```python
cv = CountVectorizer(ngram_range=(2, 2))
X = cv.fit_transform(documents)

print("Bi-gram Vocabulary:", cv.get_feature_names_out())
print("Bi-gram Matrix:\n",X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x = tfidf.fit_transform(documents)

print("Vocabulary:", tfidf.get_feature_names_out())
print("tfidf Matrix:\n", X.toarray())
```
